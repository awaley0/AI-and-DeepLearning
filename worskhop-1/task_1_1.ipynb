```python
from PIL import Image
import numpy as np
image = Image.open("/content/cat.jpg")
```

```python
display(image)
```

```python
print(image.format)
print(image.size)
print(image.mode)
```

```python
width, height = image.size
channels = len(image.getbands())
print(f"Image shape (RGB): ({height}, {width}, {channels})")
image_size = width * height * 3
print(f"Image Size (RGB) : {image_size}")
```

```python
image_array = np.array(image)
print("Shape of image array: ", image_array.shape)
print("Size of image array: ", image_array.size)

red_channel = image_array[:, :, 0]
print(red_channel)
display(red_channel)

green_channel = image_array[:, :, 1]
print(green_channel)
display(green_channel)

blue_channel = image_array[:, :, 2]
print(blue_channel)
display(blue_channel)
```
